```python
class ClusterVariants:
    def __init__(self,measurements, alg_list, stat=min):
        self.measurements = {}
        self.alg_list = alg_list
        for alg in alg_list:
            t_alg = measurements[measurements['case:concept:name'].str.contains(alg)]
            self.measurements[alg] = self.remove_outliers(list(t_alg['case:duration']))
            #self.measurements[alg] = list(t_alg['case:duration'])
                
        self.comparision_matrix = {}
        self.init_comparision_matrix()
        self.stat = stat
        
    def remove_outliers(self, measurements):
        x = np.array(measurements)
        q90, q10 = np.percentile(x,[90,10])
        max_l = q90+(1.1*(q90-q10))
        min_l = q10-(1.1*(q90-q10))
        return list(x[(x<max_l) & (x>min_l)])  
    
    def init_comparision_matrix(self):
        self.comparision_matrix = {}
        for alg in alg_list:
            self.comparision_matrix[alg] = {}
            for alg2 in alg_list:
                self.comparision_matrix[alg][alg2] = -1
        
    
    def get_measurements(self,alg):
        return self.measurements[alg]
    
    def get_quartiles(self, measurements, q_max=75, q_min=25 ):
        return np.percentile(measurements,[q_max,q_min])
    
    def compareAlgsQuartiles(self, alg1, alg2, q_max, q_min):
        if self.comparision_matrix[alg1][alg2] != -1:
            return self.comparision_matrix[alg1][alg2]
        
        t_alg1 = self.get_measurements(alg1)
        t_alg2 = self.get_measurements(alg2)
        
        q1_max, q1_min = self.get_quartiles(t_alg1, q_max, q_min)
        q2_max, q2_min = self.get_quartiles(t_alg2, q_max, q_min)

        
        ret = 1 #alg1 ~ alg2
        if q1_max < q2_min:
            ret = 0 #alg1 is faster than alg2
        elif q2_max < q1_min:
            ret = 2 #alg2 is faster than alg1 
            
        self.comparision_matrix[alg1][alg2] = ret
        self.comparision_matrix[alg2][alg1] = ret

        return ret
        
        
       
    def compareAlgs(self,alg1, alg2, threshold, M, K):
        
        #Check if comparision already exists
        if self.comparision_matrix[alg1][alg2] != -1:
            return self.comparision_matrix[alg1][alg2]
        
        t_alg1 = self.get_measurements(alg1)
        t_alg2 = self.get_measurements(alg2)

        c=0

        for m in range(M):
            m_alg1 = random.sample(t_alg1,K)
            e_alg1 = self.stat(m_alg1)
            
            m_alg2 = random.sample(t_alg2,K)
            e_alg2 = self.stat(m_alg2)
            
            
            if(e_alg1 < e_alg2):
                c = c+1
            elif(e_alg1 == e_alg2):
                c = c+0.5
                
        p = float(c)/float(M)
        #print(alg1,alg2,p)

        ret = 1 #alg1 ~ alg2
        if(p>=threshold):
            ret = 0 #alg1 is faster than alg2
        elif(p<(1-threshold)):
            ret = 2 #alg2 is faster than alg1  
            
        self.comparision_matrix[alg1][alg2] = ret
        self.comparision_matrix[alg2][alg1] = ret

        return ret
    
    def sortAlgs(self, q_max, q_min):
        
        self.init_comparision_matrix()
        
        p = len(self.alg_list)
        
        r = np.array([i for i in range(p)])
        s = np.array([i for i in range(p)])
        
        algs = {}
        for i in range(p):
            algs[s[i]] = self.alg_list[i]
        
        for i in range(p):
            for j in range(0,p-i-1):
                
                #ret = self.compareAlgs(algs[s[j]], algs[s[j+1]], threshold, M, K)
                ret = self.compareAlgsQuartiles(algs[s[j]], algs[s[j+1]], q_max, q_min)
                
                #if alg j+1 is faster than alg j
                if ret == 2: 
                    # swap index
                    s[j], s[j+1] = s[j+1], s[j]
                    
                    #update rank
                    if r[j+1] == r[j]:
                        if j==0:
                            r[j+1:] = r[j+1:]+1
                        elif r[j-1] != r[j]:
                            r[j+1:] = r[j+1:]+1
                            
                    else:
                        if j!=0 and r[j-1]!=r[j]:
                            r[j+1:] = r[j+1:] - 1
                            
                
                # alg j+1 is as good as alg j
                if ret == 1:
                    #update rank
                    if r[j+1] != r[j]:
                        r[j+1:] = r[j+1:] - 1
                            
            
            columns = ['case:concept:name', 'case:rank']
            
            return pd.DataFrame([(algs[s[i]],r[i]) for i in range(p)], columns=columns)
        
    
    def getCluster(self,rep=30, threshold=0.95, M=30, K=3):
        algs_best = []
        algs_worst = []
        
        best_class = []
        worst_class = []
        
        for i in range(rep):
            random.shuffle(self.alg_list)
            ranks = self.sortAlgs(threshold, M, K)
            
            #the best and the worst ranks
            first = 0
            last = ranks['case:rank'].max()
            
            algs_best += list(ranks[ranks['case:rank']==first]['case:concept:name'])
            if first != last:
                algs_worst += list(ranks[ranks['case:rank']==last]['case:concept:name'])
            
        for f in np.unique(algs_best):
            best_class.append((f, float(algs_best.count(f))/rep))
            
        for l in np.unique(algs_worst):
            worst_class.append((l, float(algs_worst.count(l))/rep))
            
        df_best = pd.DataFrame(best_class,columns=['case:concept:name', 'case:p-best'])
        df_worst = pd.DataFrame(worst_class,columns=['case:concept:name', 'case:p-worst'])
        df = df_best.merge(df_worst, on='case:concept:name', how="outer")
        df.fillna(0.0, inplace=True)
            
        return df
    
    def show_measurement_histograms(self, alg_list = None, bins=10):
        if not alg_list:
            alg_list = self.alg_list
        alg_list.sort()
        
        n = len(alg_list)
        fig = plt.figure(figsize=(7,3*n))
        gs = gridspec.GridSpec(n, 1, height_ratios=[1]*n) 
        
        ax = [None]*n
        for i in range(n):
            if i != 0:
                ax[i] = plt.subplot(gs[i], sharex=ax[0])
            else:
                ax[i] = plt.subplot(gs[i])
            ax[i].set_title(alg_list[i])
            ax[i].hist(self.measurements[alg_list[i]], bins=bins)
            ax[i].xaxis.set_major_formatter(FormatStrFormatter('%.e'))
            
        plt.subplots_adjust(hspace=0.5)
        plt.show()
            
        
```
